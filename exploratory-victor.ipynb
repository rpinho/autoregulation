```python
def plot_access_vs_samples_vs_noise(sampling=myrun_noiseN4s, k=4, devo_time=200, bits=4):
    for fname, color in zip(sampling, colors[None]):
        myrun = cPickle.load(open(logs_dir + fname + '.run'))
        filename = get_replaced_filename(myrun=fname, dim=k)
        #filename = get_replaced_filename(filename, samples=1e6)
        filename = get_replaced_filename(filename, myrun, bits)
        filename = get_replaced_filename(filename, devo_time=devo_time)
        data = ma.masked_invalid(loadtxt(filename + '.tsv', int))

        # mask the first point of run #2 to avoid connecting straight lines
        i = where(data.T[0] == 1)[0]
        if i.size > 1:
            data[i[-1]].mask = [True, True]

        plot(data.T[0], data.T[1], color, label=myrun.noise)

    # set_plot_kwargs
    xscale('log'); xlabel('n_samples'); ylabel('n_unique_phenotypes')
    myrun.dim = k
    title(get_myrun_title(myrun, experiment='dims overrepresentation'))
    legend(loc=0)
```
